# Handling Column Name Conflicts with suffixes=()

If both DataFrames have a column with the same name (but not part of the join), merge() will auto-rename them to avoid conflicts.

You can control that using the suffixes argument.

**suffixes=('_left', '_right')**


```python
import pandas as pd

df1 = pd.DataFrame({
    'EmpID': [101, 102],
    'Name': ['Alice', 'Bob'],
    'Dept': ['HR', 'Sales']
})

df2 = pd.DataFrame({
    'EmpID': [101, 102],
    'Name': ['Alicia', 'Bobby'],
    'Salary': [70000, 80000]
})
```

```python
df1
```

```python
df2
```

```python
# Merging with conflicting 'Name' column without suffix
merged = pd.merge(df1, df2, on='EmpID')
merged
```

```python
# Merging with conflicting 'Name' column with suffix
merged = pd.merge(df1, df2, on='EmpID', suffixes=('_Info', '_Pay'))
merged
```

# 7. Use indicator=True to track matched/unmatched rows

**When doing a merge, you may want to know:**

Did this row come from left?

From right?

Or both?

Pandas can add a column called _merge that tells you this info.

**When is this useful?**

Auditing mismatches between datasets

Debugging incorrect joins

Filtering only unmatched rows

```python
import pandas as pd

df1 = pd.DataFrame({
    'EmpID': [101, 102, 103],
    'Name': ['Alice', 'Bob', 'Charlie']
})

df2 = pd.DataFrame({
    'EmpID': [102, 104],
    'Salary': [70000, 80000]
})

# Outer merge with indicator
merged = pd.merge(df1, df2, on='EmpID', how='outer', indicator=True)
merged
```

```python
# filtering left_onlt merges
merged.loc[merged['_merge'] == 'left_only']
```

# left_on and right_on

###  Regex Cheat Sheet (for Pandas `.str`)

| Pattern   | Meaning                              | Example Match                             |
|-----------|--------------------------------------|-------------------------------------------|
| `\d`      | A digit (09)                        | `0`, `1`, ..., `9`                         |
| `\D`      | Not a digit                          | `A`, `b`, `-`, `@`                         |
| `.`       | Any single character                 | `a`, `1`, `!`                              |
| `^`       | Start of string                      | `^Hi`  matches "Hi there"                |
| `$`       | End of string                        | `\d$`  matches strings ending in a digit |
| `*`       | 0 or more of the previous thing      | `a*`  `""`, `a`, `aaa`                   |
| `+`       | 1 or more of the previous thing      | `a+`  `a`, `aa` (not `""`)               |
| `?`       | 0 or 1 of previous thing (optional)  | `a?`  `""`, `a`                          |
| `[abc]`   | Any one of `a`, `b`, or `c`          | Matches `a`, `b`, or `c`                  |
| `[^abc]`  | Anything except `a`, `b`, `c`        | Matches `d`, `x`, etc.                    |
| `{n}`     | Exactly n times                     | `\d{3}`  `123`, `456`                    |
| `()`      | Capture group (for `.str.extract()`) | `(\d+)`  captures digits like `190`     |

#  Pandas Index-based Min/Max Functions

```python
import pandas as pd

df = pd.DataFrame({
    'group': ['A', 'A', 'B', 'B'],
    'value': [3, 1, 7, 4]
})

df
```

```python
#  idxmin()  Get row with min value per group
df.loc[df.groupby('group')['value'].idxmin()]
```

```python
#  idxmax()  Get row with max value per group
df.loc[df.groupby('group')['value'].idxmax()]
```

```python
#  min()  Get min value per group
df.groupby('group')['value'].min().reset_index()
```

```python
#  max()  Get max value per group
df.groupby('group')['value'].max().reset_index()
```

```python
#  argmin()  Get position of min value (not index)
np.argmin(df['value'].values)  # returns position (e.g., 1)
df.iloc[np.argmin(df['value'].values)]  # get row at that position
```

```python
#  argmax()  Get position of max value
np.argmax(df['value'].values)
df.iloc[np.argmax(df['value'].values)]
```

```python
#  Bonus  Get index and actual value of min and max
df['value'].idxmin(), df['value'].min()
```

```python
df['value'].idxmax(), df['value'].max()
```

# Data Cleaning

| **Step**                         | **Action**                                                              | **Why**                                                    |
| -------------------------------- | ----------------------------------------------------------------------- | ---------------------------------------------------------- |
| 1. Understand the Data           | `df.head()`, `df.info()`, `df.describe(include='all')`                  | Understand structure, types, and missing or unusual values |
| 2. Rename & Standardize Columns  | `df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')` | Make column names consistent and easier to reference       |
| 3. Standardize Strings           | `.str.strip()`, `.str.lower()`, `.str.replace()`                        | Clean up inconsistent capitalization, spaces, typos, etc.  |
| 4. Handle Duplicates             | `df.drop_duplicates()`                                                  | Remove exact row repeats that can corrupt analysis         |
| 5. Handle Invalid/Extreme Values | e.g. `df.loc[df['age'] > 120, 'age'] = np.nan`                          | Replace unrealistic or corrupt data                        |
| 6. Handle Missing Values         | `df.dropna()`, `df.fillna()`, `df.ffill()`, `groupby().transform()`     | Deal with nulls depending on the context and importance    |
| 7. Fix Data Types                | `pd.to_datetime()`, `pd.to_numeric()`, `.astype()`                      | Ensure correct data types for operations and modeling      |
| 8. Encode Categorical Variables  | `.map()`, `pd.get_dummies()`, `.astype('category')`                     | Convert text categories into usable numeric formats        |
| 9. Feature Engineering           | Extract from date, calculate ratios, flags, etc.                        | Add new variables that can improve insights or models      |
| 10. Final Checks & Cleanup       | `df.isna().sum()`, `df.dtypes`, `df.reset_index(drop=True)`             | Ensure clean, ready-to-analyze or model dataset            |

# String standardization

| **Category**           | **Method**                            | **Description**                            | **Example Output**                  |
| ---------------------- | ------------------------------------- | ------------------------------------------ | ----------------------------------- |
|  Case Handling       | `str.lower()`                         | Lowercase all characters                   | `"HELLO"`  `"hello"`               |
|                        | `str.upper()`                         | Uppercase all characters                   | `"hello"`  `"HELLO"`               |
|                        | `str.title()`                         | Title case (First Letter Capitalized)      | `"hello world"`  `"Hello World"`   |
|                        | `str.capitalize()`                    | Capitalize only first word                 | `"hello world"`  `"Hello world"`   |
|                        | `str.swapcase()`                      | Swap upper to lower and vice versa         | `"Hello"`  `"hELLO"`               |
|  Whitespace Handling | `str.strip()`                         | Remove leading/trailing whitespace         | `" hi "`  `"hi"`                   |
|                        | `str.lstrip()`                        | Remove leading whitespace                  | `" hi "`  `"hi "`                  |
|                        | `str.rstrip()`                        | Remove trailing whitespace                 | `" hi "`  `" hi"`                  |
|                        | `str.pad(width=10)`                   | Pad strings with spaces                    | `"hi"`  `"       hi"`              |
|  Replace & Remove    | `str.replace('old', 'new')`           | Replace substrings                         | `"abc".replace("b", "x")`  `"axc"` |
|                        | `str.replace(r'\d+', '', regex=True)` | Remove all digits via regex                | `"ab123c"`  `"abc"`                |
|                        | `str.replace(' ', '_')`               | Replace space with underscore              | `"hello world"`  `"hello_world"`   |
|  Searching & Match   | `str.contains('word')`                | Check if string contains word              | `"cat"`  `True` for `"bobcat"`     |
|                        | `str.startswith('pre')`               | Check if string starts with                | `"data.csv"`  `True`               |
|                        | `str.endswith('.csv')`                | Check if string ends with                  | `"data.csv"`  `True`               |
|  Slicing & Splitting | `str[:5]`                             | Get first 5 characters                     | `"abcdef"`\[:5]  `"abcde"`         |
|                        | `str.slice(2, 5)`                     | Slice from position 2 to 4                 | `"abcdef"`  `"cde"`                |
|                        | `str.split(' ')`                      | Split into list by delimiter               | `"a b"`  `["a", "b"]`              |
|                        | `str.split(' ', expand=True)`         | Split into multiple columns                | `"a b"`  2 columns `"a"`, `"b"`    |
|                        | `str.extract(r'(\d+)')`               | Extract regex match (e.g., numbers)        | `"abc123"`  `"123"`                |
|  Length & Counts     | `str.len()`                           | Length of each string                      | `"abc"`  `3`                       |
|                        | `str.count('a')`                      | Count number of occurrences of a character | `"banana"`  `3`                    |
|  Advanced Tips       | `str.findall(r'\d+')`                 | Find all digit substrings                  | `"abc123de4"`  `["123", "4"]`      |
|                        | `str.zfill(5)`                        | Pad with leading zeros to length 5         | `"42"`  `"00042"`                  |


```python
df['name'] = df['name'].str.strip().str.title()   # Trim spaces, title-case names
df['department'] = df['department'].str.strip().str.title()
df
```

```python
# observe that age column has -1 and 120 like some outliers, so converting the age column to numeric datatype and flagging few unrealistic values by null values
df['age'] = pd.to_numeric(df['age'], errors='coerce')

# Flag and Replace Unrealistic Ages with NaN
df.loc[(df['age'] < 18) | (df['age'] > 100), 'age'] = np.nan
df
```

```python
# standardizing gender
df['gender'] = df['gender'].str.upper().str.strip()  # normalize if not already
df['gender'] = df['gender'].replace({'MALE': 'M', 'FEMALE': 'F'})  # standardize

# Step 3: fill all unknown or missing values with 'other'
df['gender'] = df['gender'].where(df['gender'].isin(['M', 'F']), 'other')
df
```

 4. Handle Missing Values

#  Pandas Missing Values Cheat Sheet

| Strategy                         | Code Example                                                   | Best Used When                                               |
|----------------------------------|----------------------------------------------------------------|--------------------------------------------------------------|
|  Drop all rows with any NaN    | `df.dropna()`                                                  | You can afford to lose rows completely                       |
|  Drop rows where `col` is NaN  | `df.dropna(subset=['col'])`                                    | Only specific column is critical                             |
|  Fill with fixed value         | `df['col'].fillna(0)` or `'Unknown'`                           | Default makes sense for business logic                       |
|  Forward fill (last known)     | `df.fillna(method='ffill')`                                    | Ordered or time-series data                                  |
|  Backward fill (next known)    | `df.fillna(method='bfill')`                                    | Ordered data with missing beginnings                        |
|  Fill with mean                | `df['col'].fillna(df['col'].mean())`                           | Numerical, symmetric distributions                           |
|  Fill with median              | `df['col'].fillna(df['col'].median())`                         | Numerical, skewed distributions                              |
|  Fill with mode                | `df['col'].fillna(df['col'].mode()[0])`                        | Categorical, frequent value makes sense                      |
|  Group-based imputation        | `df.groupby('group')['col'].transform(lambda x: x.fillna(x.mean()))` | When values depend on groups (e.g. gender-wise age avg) |
|  Custom rule-based filling     | `df.apply(lambda row: ..., axis=1)`                            | Logic depends on multiple columns                            |

 Always check:  
- `df.isna().sum()` before and after  
- `df.dtypes` if conversions were done



 Detecting Missing Values

| Method         | Description            | Example                 |
| -------------- | ---------------------- | ----------------------- |
| `df.isna()`    | Detect `NaN` values    | `df[df['age'].isna()]`  |
| `df.isnull()`  | Same as `isna()`       | `df['age'].isnull()`    |
| `df.notna()`   | Detect non-null values | `df[df['age'].notna()]` |
| `df.notnull()` | Same as `notna()`      | `df['age'].notnull()`   |

 Dropping Missing Values

| Method                      | Description                                     |
| --------------------------- | ----------------------------------------------- |
| `df.dropna()`               | Drop rows with any `NaN`                        |
| `df.dropna(axis=1)`         | Drop columns with any `NaN`                     |
| `df.dropna(how='all')`      | Drop rows/cols only if **all** values are `NaN` |
| `df.dropna(subset=['col'])` | Drop rows if `col` is `NaN`                     |


 Filling Missing Values

| Method                               | Description                       |
| ------------------------------------ | --------------------------------- |
| `df.fillna(0)`                       | Fill all `NaN` with 0             |
| `df['col'].fillna(df['col'].mean())` | Fill with column mean             |
| `df['col'].fillna(method='ffill')`   | Forward fill (use previous value) |
| `df['col'].fillna(method='bfill')`   | Backward fill (use next value)    |
| `df['col'].ffill().bfill()`          | Forward + backward fill combo     |


 Interpolating Values

| Method                      | Description                        |
| --------------------------- | ---------------------------------- |
| `df['value'].interpolate()` | Fill numeric gaps by interpolation |


 Replacing Custom Missing Indicators

| Example                                             | Description              |
| --------------------------------------------------- | ------------------------ |
| `df.replace(['N/A', 'invalid', 'Unknown'], np.nan)` | Convert strings to `NaN` |



```python
df.isna().sum()             # Check missing values
```

```python
df['name'].fillna('Unknown', inplace=True) # filling NULL with Unknown for name field
```

```python
# replacing null age values by mean of age by gender
avg_age_by_gender = df.groupby('gender')['age'].mean().round().astype('Int64')  # Using 'Int64' allows nullable integers
# Step 2: Fill missing ages using map from gender
df['age'] = df.apply(
    lambda row: avg_age_by_gender[row['gender']] if pd.isna(row['age']) else row['age'],
    axis=1
)

df
```

```python
# taking salary avg by department for null values
avg_salary_by_dept = df.groupby('department')['salary'].mean().round()

# Step 2: Fill missing salary using map from dept
df['salary'] = df.apply(
    lambda row: avg_salary_by_dept[row['department']] if pd.isna(row['salary']) else row['salary'],
    axis=1
)
df
```

#  Pandas Duplicate Handling  Complete Cheat Sheet

| **Category**           | **Method / Command**                          | **Description**                                                                  | **Example / Notes**                                 |
| ---------------------- | --------------------------------------------- | -------------------------------------------------------------------------------- | --------------------------------------------------- |
|  Check Duplicates    | `df.duplicated()`                             | Returns a Boolean Series: `True` if row is duplicate (first occurrence is False) | `df[df.duplicated()]` shows all duplicates          |
|                        | `df.duplicated(subset=['col1', 'col2'])`      | Check for duplicates based on specific columns                                   | Useful when only some fields matter                 |
|                        | `df.duplicated(keep='first')`                 | Mark duplicates except for **first** occurrence (default)                        | `keep='last'`, `keep=False` are other options       |
|  Count Duplicates    | `df.duplicated().sum()`                       | Number of duplicate rows                                                         |                                                     |
|  Drop Duplicates    | `df.drop_duplicates()`                        | Drop **duplicate rows** based on all columns                                     | Keeps the **first** by default                      |
|                        | `df.drop_duplicates(subset=['col'])`          | Drop duplicates using selected column(s) only                                    | Often used in deduping based on `email`, `id`, etc. |
|                        | `df.drop_duplicates(keep='last')`             | Keep **last** occurrence instead of first                                        |                                                     |
|                        | `df.drop_duplicates(keep=False)`              | Drop **all** duplicates (keep none)                                              |                                                     |
|  Reset Index After   | `df.drop_duplicates().reset_index(drop=True)` | Resets index after dropping duplicates                                           | Optional but recommended                            |
|  Get Only Duplicates | `df[df.duplicated(keep=False)]`               | Show all **duplicate** rows including both copies                                |                                                     |
|  Marking Duplicates | `df['is_duplicate'] = df.duplicated()`        | Add a flag column indicating duplicates                                          | Good for debugging or review                        |

#  Categorical Variable Encoding  Pandas Cheat Sheet

| **Technique**              | **Command / Method**                                                  | **Description**                                                      | **Example Output**                            |
| -------------------------- | --------------------------------------------------------------------- | -------------------------------------------------------------------- | --------------------------------------------- |
|  **Label Encoding**      | `df['col'].astype('category').cat.codes`                              | Assigns a unique number to each category                             | `"Male"  0`, `"Female"  1`                  |
|                          | `from sklearn.preprocessing import LabelEncoder`                      | Use scikit-learn's version (if needed)                               | `le.fit_transform(df['col'])`                 |
|  **One-Hot Encoding**     | `pd.get_dummies(df, columns=['col'])`                                 | Creates binary columns for each category                             | `"Gender_Male", "Gender_Female"`              |
|                          | `pd.get_dummies(df['col'])`                                           | Use if you only want dummy for one column                            |                                               |
|  Drop first dummy        | `pd.get_dummies(df, drop_first=True)`                                 | Avoid multicollinearity (useful in modeling)                         | Drops one column from each one-hot group      |
|  **Binary Encoding**     | `df['col'].map({'Yes': 1, 'No': 0})`                                  | Manually encode binary labels                                        | `"Yes"  1`, `"No"  0`                       |
|  **Ordinal Encoding**    | `df['col'] = df['col'].map({'Low': 1, 'Medium': 2, 'High': 3})`       | Encode meaningful order                                              | Preserves hierarchy: Low < Medium < High      |
|  **Frequency Encoding**  | `df['col'].map(df['col'].value_counts())`                             | Replaces categories with their frequency count                       | `"A"  12`, `"B"  4`, etc.                   |
|  Combine Rare Categories | `df['col'] = df['col'].apply(lambda x: x if x in top_5 else 'Other')` | Combines infrequent classes                                          | Reduces sparsity in high-cardinality features |
|  **Target Encoding**     | `df.groupby('col')['target'].mean()`  then `.map()`                  | Replaces category with mean of target variable (risk: data leakage!) | Useful in ML pipelines                        |


** Tips for Choosing Encoding**

| Use Case                                      | Recommended Technique        |
| --------------------------------------------- | ---------------------------- |
| Binary variables                              | Binary encoding / map()      |
| Nominal (no order, few values)                | One-Hot Encoding             |
| Ordinal (ordered categories)                  | Ordinal Encoding / map()     |
| High cardinality categorical (e.g. zip codes) | Frequency or Target encoding |
